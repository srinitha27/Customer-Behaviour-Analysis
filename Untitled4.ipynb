```python
import pandas as pd
```

```python
df = pd.read_csv('/content/superstore_final_dataset (1).csv', encoding='latin1')
display(df.head())
```

```python
df.info()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d/%m/%Y')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format='%d/%m/%Y')

print(f"Number of duplicate rows before dropping: {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
print(f"Number of duplicate rows after dropping: {df.duplicated().sum()}")

df.info()
```

```python
# Calculate Revenue
revenue = df['Sales'].sum()
print(f"Total Revenue: ${revenue:,.2f}")

# Calculate Average Order Value (AOV)
aov = df['Sales'].sum() / df['Order_ID'].nunique()
print(f"Average Order Value (AOV): ${aov:,.2f}")
```

```python
print("Missing values before handling:")
display(df.isnull().sum())
```

```python
df['Postal_Code'].fillna(0, inplace=True)
print("Missing values after handling:")
display(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Order_Month'] = df['Order_Date'].dt.to_period('M')
monthly_sales = df.groupby('Order_Month')['Sales'].sum().reset_index()
monthly_sales['Order_Month'] = monthly_sales['Order_Month'].astype(str)

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Order_Month', y='Sales')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_categories = df.groupby('Category')['Sales'].sum().nlargest(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=top_categories, x='Sales', y='Category')
plt.title('Top 10 Product Categories by Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('Product Category')
plt.tight_layout()
plt.show()
```

```python
top_sub_categories = df.groupby('Sub_Category')['Sales'].sum().nlargest(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=top_sub_categories, x='Sales', y='Sub_Category')
plt.title('Top 10 Product Sub-Categories by Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('Product Sub-Category')
plt.tight_layout()
plt.show()
```

```python
regional_sales = df.groupby('Region')['Sales'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=regional_sales, x='Region', y='Sales')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.show()
```
